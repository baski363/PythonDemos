```python
# int() truncates a number to an integer
print(
    int(2.3),
    int(2.7),
    int(-2.3),
    int(-2.7)
)
```

```python
# function that checks if i is even
def is_even(i):
    if i/2==int(i/2):
        return 1       # yes
    else:
        return 0       # no
    
print(is_even(3))
print(is_even(4))
print(is_even(4.6))
```

```python
# example with two arguments
def find_product_of_a_and_b(a,b):
    return a*b

print(find_product_of_a_and_b(3,5))
```

```python
# example with two outputs
def find_sum_and_then_product(a,b):
    return a+b,a*b

print(find_sum_and_then_product(3,5))
```
