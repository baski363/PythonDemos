# Matplotlib

We are going to be making some plots in this exercise.  Before we can do so we need to import the libraries to do so.  In the cell below, use the `%pylab` command to import Matplotlib and Numpy for inline plots.

```python
%pylab inline --no-import-all
```

```python
assert np.sin(0) == 0.
```

First let's make up some data for us to plot up.  Using Numpy's `linspace` command generate an array called `t` that will start at zero and go to 7 (the default number of data points is fine).  For `y` values we will use `sin(t)`.

```python
t = np.linspace(0, 7)
y = np.sin(t)
```

```python
assert y[24] == -0.28305585408225559
```

Make a plot of `y` vs. `t` where `t` is on the horizontal axis and `y` is on the vertical.  Label the `t` axis "Time, s" and the `y` axis "Voltage, V".  

```python
plt.plot(t, y)
plt.xlabel('Time, s')
plt.ylabel('Voltage, V')
```

Now, let's add a delayed version of the `y` values called `y_delayed` such that $y_{delayed} = \sin(t + \pi/2)$.  In the cell below create a plot with both `y` and `y_delayed` on it and include a legend.

```python
y_delayed = np.sin(t + np.pi/2)

plt.plot(t, y, label = 'Original')
plt.plot(t, y_delayed, label = 'Delayed')
plt.xlabel('Time, s')
plt.ylabel('Voltage, V')
plt.legend(loc = 'lower left')
```

# Plotting Styles

In this section we want to explore the built-in styles that are available.  Print out the styles that are available on your computer.

```python
plt.style.available
```

Using the third style listed, re-create the last plot in the previous section.

```python
with plt.style.context('bmh'):
    plt.plot(t, y, label = 'Original')
    plt.plot(t, y_delayed, label = 'Delayed')
    plt.xlabel('Time, s')
    plt.ylabel('Voltage, V')
    plt.legend(loc = 'lower left')
```

Change the plot so that the y limits are -1.05 and 1.05.  

Now create a new plot of the same data that has a size of 8 wide and 6 high.  Also use a font size of 12 for the x- and y-ticks, a font size of 14 for the x- and y-labels and a font size of 14 for the legend entries.

```python
with plt.style.context('bmh'):
    plt.figure(figsize = (8, 6))
    plt.plot(t, y, label = 'Original')
    plt.plot(t, y_delayed, label = 'Delayed')
    plt.xlabel('Time, s', fontsize = 14)
    plt.ylabel('Voltage, V', fontsize = 14)
    plt.xticks(fontsize = 12)
    plt.yticks(fontsize = 12)
    plt.ylim([-1.05, 1.05])
    plt.legend(loc = 'lower left', fontsize = 14)
```
